# Testing implementation for hash method

```python
key = 'kevin'
letters = list(key)
print(letters)
```

```python
unicodes = list(ord(letter) for letter in letters)

print(unicodes)
hash_value = sum(unicodes)
```

```python
collection = {}
collection[hash_value] = {"kevin":"person"}
print(collection)
collection[hash_value] = {"Alice":"person"}
print(collection)

```
