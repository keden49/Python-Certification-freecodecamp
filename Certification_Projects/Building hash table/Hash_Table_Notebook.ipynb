```python
class HashTable:
    def __init__(self): #initialize an empty dictionary
        self.collection = {} #stores hash_value : {key:value}

    # Creating Hash function

    def hash(self,key=str):
        letters = list(key)#key should be in string value
        unicodes = list(ord(letter)for letter in letters)

        return sum(unicodes)

    # Creating add method

    def add(self,key,value):
        hash_value = self.hash(key)

        if hash_value in self.collection:
            self.collection[hash_value][key] = value

        else:
            self.collection[hash_value] = {key:value}


    #Creating remove method

    def remove(self,key):
        hash_value = self.hash(key)

        if hash_value in self.collection:
          if key in self.collection[hash_value]:#Test 10
            '''
            deletes the specific key-value pair in nested dictonaries with more than one value'''

            if len(self.collection[hash_value])> 1:
              del self.collection[hash_value][key] #Test 11
            else:
              del self.collection[hash_value]
          else:
            pass #returns nothing when key does not exist

    def lookup(self,key):
        hash_value = self.hash(key)
        if hash_value in self.collection:
            return self.collection[hash_value].get(key,None)
        else:

            return None
```
