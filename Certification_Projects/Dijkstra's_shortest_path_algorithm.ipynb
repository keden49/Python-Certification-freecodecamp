# Notebook Overview

In this workshop, you will implement the shortest path algorithm. You will write a Python function that computes the shortest path between the nodes in a graph, and also returns the path taken.

For example, given a graph where cities are connected by roads with different distances, the algorithm will find the shortest route from one city to another. If you want to travel from City A to City D, the algorithm might find that going A ⇨ B ⇨ C ⇨ D (total: 15km) is shorter than going directly A ⇨ D (20km).

```python
INF = float('inf')
adj_matrix = [
    [0, 5, 3, INF, 11, INF],
    [5, 0, 1, INF, INF, 2],
    [3, 1, 0, 1, 5, INF],
    [INF, INF, 1, 0, 9, 3],
    [11, INF, 5, 9, 0, INF],
    [INF, 2, INF, 3, INF, 0],
]

def shortest_path(matrix, start_node, target_node=None):
    n = len(matrix) #number of nodes
    distances = [INF] * n#initializing all nodes as infinity / temporary state
    distances[start_node] = 0 #setting start node as 0
    paths = [[node_no] for node_no in range(n)] #stores shortest paths to each node from start node
    visited = [False] * n #prevents processing nodes twice

    '''
    adds loop to visited list
    a node is marked as visited when it is selected as the node with
    the smallest tentative distance from the unvisited set
    and the algorithm has finished examining all of its neighbors
    '''


    for _ in range(n):
        min_distance = INF
        current = -1

        '''
        Greedy as it selects the most minimum path
        and pursues that node

        '''

        for node_no in range(n): #determines next node to be pursued
            if not visited[node_no] and distances[node_no] < min_distance:
                min_distance = distances[node_no]
                current = node_no

        if current == -1: #incase no connection is in between
            break
        visited[current] = True

        for node_no in range(n): #visits neighbours of current node
            distance = matrix[current][node_no] # find distance from current node
            if distance != INF and not visited[node_no]:
               #distances[current] factors in preceding nodes
                new_distance = distances[current] + distance
                if new_distance < distances[node_no]: #reassigning INF value/ relaxing distance
                    distances[node_no] = new_distance #reassignment is only done when current visited nodes neighbours distances are less than what is currently held
                    paths[node_no] = paths[current] + [node_no]





    targets = [target_node] if target_node is not None else range(n)
    for node_no in targets:
        if node_no == start_node or distances[node_no] == INF:
            continue
        string_path = (str(n) for n in paths[node_no])
        path = ' -> '.join(string_path)
        print(f'\n{start_node}-{node_no} distance: {distances[node_no]}\nPath: {path}')

    return distances, paths

shortest_path(adj_matrix,0,5)

```

# Conclusion


1. Find the most minimum path and pursue it
2. Mark that path/node as visited
3. Visit all current nodes unvisited neighbours
4. Relax its current distances
5. Only make change to its distance when a much shorter distance is found than what is currently present
6. Repeat the process
