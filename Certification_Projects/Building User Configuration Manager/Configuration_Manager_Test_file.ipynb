```python
def add_setting(settings,key_value):
    #convert to lowercase
    key,value = key_value
    key = key.lower()
    value = value.lower()
    if key in settings.keys():
        return f"Setting '{key}' already exists! Cannot add a new setting with this name."
    else:
        settings[key] = value
        return f"Setting '{key}' added with value '{value}' successfully!"
```

```python
def update_setting(settings,key_value):
    #convert to lowercase
    key,value = key_value
    key = key.lower()
    value = value.lower()
    if key in settings.keys():
        settings[key]=value
        return f"Setting '{key}' updated to '{value}' successfully!"
    else:
        return f"Setting '{key}' does not exist! Cannot update a non-existing setting."

```

```python
#TEST 13

print(update_setting({'theme': 'light'}, ('theme', 'dark')))

# TEST 14

print(update_setting({'theme': 'light'}, ('volume', 'high')))
```

```python
def delete_setting(settings,key_value):
    #convert to lowercase

    key = key_value.lower()
    if key in settings.keys():
        del settings[key]
        return f"Setting '{key}' deleted successfully!"
    else:
        return "Setting not found!"
```

```python
def view_settings(settings):
    if settings == {}:
        return "No settings available."
    else:
        output ="Current User Settings:"

        for key,value in settings.items():

          output+= f"\n{key.capitalize()}: {value}"

        return output + "\n"
```

# Lesson from view setting

1. Return ends an operation
2. Hence when called inside a loop it only returns first item
3. Return can return other attributes such as string item
4. To add items to a string use += operator
5. In a for loop items will be added one by one
6. To begin pair values in a new line use \n at the beggining
7. To end with a new_line character without looping append outside the loop

```python
#TEST 24
settings = {}
print(view_settings(settings))

#TEST 25-27

formatted_settings= view_settings({'theme': 'dark', 'notifications': 'enabled', 'volume': 'high'})
print(formatted_settings)
```
