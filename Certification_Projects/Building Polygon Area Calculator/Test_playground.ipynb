# Tetsing rectangle string method

```python
from math import sqrt
class Rectangle:
    # creating constructor

    def __init__ (self,width,height):
        self.width = width
        self.height = height

    #width setter

    def set_width(self,value):
        self.width = value

    #height setter
    def set_height(self,value):
        self.height = value

    #Get Area
    def get_area(self):
        return self.width * self.height

    #Get Perimeter
    def get_perimeter(self):
        return 2 * (self.width + self.height)

    #Get diagonal

    def get_diagonal(self):
        square_diagonal = self.width**2 + self.height**2
        return sqrt(square_diagonal)
    #Get shape
    def get_picture(self):
        shape = ""
        if self.width > 50 or self.height >50:
            return "Too big for picture."
        else:
            for _ in range(self.height):
                shape += "*" * (self.width) + "\n"
            return shape


    #Get amount of shape that fits inside

    def get_amount_inside(self,shape):
        Area = self.get_area()
        area = shape.get_area()

        return Area//area
    #string output
    def __str__ (self):
        return f"Rectangle(width={self.width}, height={self.height})"


```

```python
rectangle = Rectangle(10,10)
print(rectangle)
```

#Testing Inheritance

```python
class Square(Rectangle):
  def __init__(self,side):
    super().__init__(side,side)#avoids duplicate work or overriding logic defined in the parent class

  def set_width(self, value):
    super().set_width(value)#avoids duplicate work
    super().set_height(value)#avoids duplicate work

  def set_height(self, value):
    super().set_height(value)
    super().set_width(value)

  def set_side(self,value):
    self.width = value
    self.height = value

  def __str__(self):
    return f"Square(side={self.width})"
```

```python
square = Square(16)
print(square.width)
square.set_side(2)
print(square)
```
