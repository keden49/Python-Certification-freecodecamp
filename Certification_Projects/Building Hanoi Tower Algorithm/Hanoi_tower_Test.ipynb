```python
def hanoi_solver(n):#Test 1,2

  rods = [list(range(1,n+1))[::-1],[],[]]
  initial_state = f"{rods[0]} {rods[1]} {rods[2]}"+"\n"

  def algorithm(disks,start,end):
    nonlocal rods

    if disks == 1:
      rods[end].append(rods[start].pop())
      return f"{rods[0]} {rods[1]} {rods[2]}"

    else:

      helper_rod = 3-(start + end)
      top_disk_move = algorithm(disks -1,start,helper_rod)
      current_disk_move = algorithm(1,start,end)
      top_disk_finalizer = algorithm(disks - 1,helper_rod,end)
      return top_disk_move +"\n" + current_disk_move + "\n" + top_disk_finalizer


  return initial_state + algorithm(n,0,2)#Test 3
```
