```python
def apply_discount(price,discount):
  #Validates data type
    if type(price) not in[int,float]:
        return 'The price should be a number'
    if type(discount) not in [int,float]:
        return 'The discount should be a number'
        #Validates price is not less than 0
    if price <= 0 :
        return 'The price should be greater than 0'
        #Validates discount range is between 0 and 100
    if  discount < 0 or discount > 100 :
        return 'The discount should be between 0 and 100'
        #gets discount
    final_discount = (discount / 100)*price

    final_price = price - final_discount
    return final_price


'''
testing area
'''
price1 = 100
discount1= 20
print(apply_discount(price1,discount1))

price2 = 200
discount2= 50
print(apply_discount(price2,discount2))

price3 = 50
discount3 = 0
print(apply_discount(price3,discount3))

price4 = 4
discount4 = 100
print(apply_discount(price4,discount4))

price5 = 74.5
discount5 = 20.0
print(apply_discount(price5,discount5))




```
