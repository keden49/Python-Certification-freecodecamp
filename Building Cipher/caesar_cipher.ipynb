```python
def caesar(text, shift, encrypt=True):

    if not isinstance(shift, int):
        return 'Shift must be an integer value.'  #Prevent crashes - Stop execution before invalid operations

    if shift < 1 or shift > 25:#wraps around the alphabet
        return 'Shift must be an integer between 1 and 25.'

    alphabet = 'abcdefghijklmnopqrstuvwxyz'

    if not encrypt:
        shift = - shift #allows for decryption

    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    #ive found a new letter way to understand it with a slicing method
    #a gap of 3(shift) is left at the beggining  each letter moves by 3(shift) to fill the gap  position
    #this looks like decryption when you view it after the slicing has occured but its encryption in the long run

    translation_table = str.maketrans(alphabet + alphabet.upper(), shifted_alphabet + shifted_alphabet.upper())
    #Takes two strings of equal length
    #Maps each character from first string to corresponding character in second string
    #Stores mapping as unicodes (97(a):110(n) )
    print(translation_table)
    encrypted_text = text.translate(translation_table)
    #Goes through each character in a string and replaces it according to the table
    return encrypted_text ## ‚Üê Sends result BACK to caller
                          ## Enables chaining - Result can be used by other functions

def encrypt(text, shift):
    return caesar(text, shift)

def decrypt(text, shift):
    return caesar(text, shift, encrypt=False)
encrypted_text = 'Pbhentr vf sbhaq va hayvxryl cynprf.'
decrypted_text= decrypt(encrypted_text,13)
print(encrypted_text)
print(decrypted_text)
```
