This Notebook uses priciples in linear structures by freecodecamp to build a linked list.

```python
class LinkedList: #Creates a list object to keep track of nodes in memory
    class Node:#actual node object, in list
        def __init__(self, element):
            self.element = element #nodes content
            self.next = None#new intialized node lacks next node since its last line

    def __init__(self):
        self.length = 0# a new instance of list zero elements/nodes
        self.head = None# head is set to none due to no nodes

    def is_empty(self):
        return self.length == 0 #returns True when list is empty

    def add(self, element): #adding elements
        node = self.Node(element)#accesses the node object in its memory and creates a new instance with it using the element passed
        if self.is_empty():
            self.head = node
        else:
            current_node = self.head #acessing the head element
            '''
            the goal of the while loop is to find the current tail
            tail - last node
            so as to reassign its pointer/next to the new node thats being added
            the new added node doesnt have any node behind it thus its next attribute is None
            intitialized during its creation
            '''

            while current_node.next is not None:
                current_node = current_node.next#reassigns the node to the node behind /travels through all the nodes
            current_node.next = node
        self.length += 1 #incrementing the items in the list


    '''
    Goal of remove is to reassign the removed node next to the node
    that was infront of it /previous node
    effectively removing of it
    '''

    def remove(self, element):
        current_node = self.head
        previous_node = None #all operation begin from the head, since the head is the first element it has no previous node
        while current_node is not None and current_node.element != element:
            previous_node = current_node #travels through node,each time it itertes, the node it was on becomes the previous node
            current_node = current_node.next#provides the next in line
        if current_node is None:
            return  #returns incase list is empty
        elif previous_node is not None:
            previous_node.next = current_node.next#effectivly reassigns the pointer of removed node to the node that was infront it
        else:
            self.head = current_node.next#handles head removal
        self.length -= 1#deducts the removed list from memory

my_list = LinkedList()
print(my_list.is_empty())

my_list.add(1)
my_list.add(2)
print(my_list.is_empty())
print(my_list.length)

my_list.remove(1)
print(my_list.length)
```
