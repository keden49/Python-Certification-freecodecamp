```python
def pin_extractor(poems):
    secret_codes = []#empty list for storing secret_codes from each poem
    for poem in poems:

        secret_code = ''
        lines = poem.split('\n')#splits columns in form of lines
        for line_index, line in enumerate(lines):#keeps track of line indexes in enumerate form
            words = line.split()#returns list of words in lines split into separate strings
            if len(words) > line_index:#makes sure the number of words in a line fit the number of lines in the poem
                                       #this is to prevent outliers eg, a line maybe index two, but they are only 2 words in that line,(index 0 & 1 )
                secret_code += str(len(words[line_index]))#takes index of line in whole poem and accesses the the word with the same index in that line
            else:                                         #takes the legth of that word, converts it to a string, adds it to empty string (empty_string)
                secret_code += '0'#converts outliers to '0'
        secret_codes.append(secret_code)   # end of operation for line,hence returned on inner loop, to allow it to move to the next line,once done it moves to the next poem
    return secret_codes                    # returned on the outer for loop after loop is complete


poem = """Stars and the moon
shine in the sky
white and
until the end of the night"""

poem2 = 'The grass is green\nhere and there\nhoping for rain\nbefore it turns yellow'
poem3 = 'There\nonce\nwas\na\ndragon'

print(pin_extractor([poem, poem2, poem3]))
```
