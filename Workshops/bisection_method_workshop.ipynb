```python
def square_root_bisection(square_target, tolerance=1e-7, max_iterations=100):
    if square_target < 0:#prevents negative values
        raise ValueError('Square root of negative number is not defined in real numbers')
    if square_target == 1:
        root = 1
        print(f'The square root of {square_target} is 1')
    elif square_target == 0:
        root = 0
        print(f'The square root of {square_target} is 0')

    else: #handles larger square targets
        low = 0
        high = max(1, square_target) #handles values that are less than 1 but greater than 0 eg 0.25
        root = None #returns None when convergence is not met

        for _ in range(max_iterations): #changes intervals as per number of iterations
            mid = (low + high) / 2
            square_mid = mid**2

            if abs(square_mid - square_target) < tolerance: #sets tolerance
                root = mid
                return root

            elif square_mid < square_target:
                low = mid
            else:
                high = mid

        if root is None:
            print(f"Failed to converge within {max_iterations} iterations.")

        else:
            print(f'The square root of {square_target} is approximately {root}')



N = 16
square_root_bisection(N)
```
