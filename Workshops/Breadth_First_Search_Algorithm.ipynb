# Notebook Overview

This workshop demonstrates BFS algorithm by building a parantheses generator which gives different combinations of parantheses based on pairs intended

It maintains properties of a BFS such as :

1. Queue data structure - First elements to a list are removed/ processed first before moving to the next level

2. Level-by-Level Exploration: The while queue: loop processes states in the order they were added. This means all partial strings of length k are fully explored and their potential next states (length k+1) are added to the queue before any strings of length k+1 are processed. This ensures a level-by-level traversal characteristic of BFS.

3. Generating Neighbors (or Next States): Inside the else block, the conditions if opens_used < pairs: and if closes_used < opens_used: define the rules for generating valid "neighbors" or subsequent states:

- If we can still add an opening parenthesis without exceeding the total pairs, a new state with an added ( is enqueued.
If we can add a closing parenthesis (only if there are more open parentheses than closed ones, to maintain validity), a new state with an added ) is enqueued.



```python
def gen_parentheses(pairs): #number of pairs intended
    if not isinstance(pairs, int):
        return 'The number of pairs should be an integer'
    if pairs < 1:
        return 'The number of pairs should be at least 1'

    queue = [('', 0, 0)] #keeps track of what should be updated next
    result = []
    while queue: #queue all possible combinations that may be born out of the pairs
        current, opens_used, closes_used = queue.pop(0)
        if len(current) == 2 * pairs:
            result.append(current)#appends a complete combination
        else:
            if opens_used < pairs: #maintains number of pairs
                queue.append((current + '(', opens_used + 1, closes_used))
            if closes_used < opens_used:
                queue.append((current + ')', opens_used, closes_used + 1))

    return result
print(gen_parentheses(3))
```
