````python
class Email:
    def __init__(self, sender, receiver, subject, body):
        self.sender = sender
        self.receiver = receiver
        self.subject = subject
        self.body = body
        self.read = False

    def mark_as_read(self):
        self.read = True

class User:
    def __init__(self, name):
        self.name = name
        self.inbox = Inbox()

    def send_email(self, receiver, subject, body):
        email = Email(sender=self, receiver=receiver, subject=subject, body=body)
        receiver.inbox.receive_email(email)

class Inbox:
    def __init__(self):
        self.emails = []

    def receive_email(self, email):
        self.emails.append(email)

    def __len__(self):
        return len(self.emails)

alice = User("Alice")
bob = User("Bob")

alice.send_email(bob, "Hello", "Hi Bob, how are you?")
print(len(bob.inbox.emails))
````

# New Fact

When you run code in Google Colab, that code isn't running on your personal computer. It's executing on a virtual machine (a kind of server) in Google's data centers. Each server has its own internal clock and is configured with a default timezone. So, datetime.datetime.now() without any timezone information will return the current time according to that server's clock and timezone settings, not necessarily your local time, or the time in any specific country you're interested in.
