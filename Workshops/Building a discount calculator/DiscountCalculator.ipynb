 # Workshop Overview
 This Workshop aims to demonstrate and apply principles in python Abstraction Object Oriented Programming to build a discount calculator which calculates the best price from the various discount strategies namely:
 ## FixedAmountDiscount
 Calculates discount by removing a fixed amount from the products price, 90% of the product price should be greater than the fixed amount

 ## PercentageDiscount
 Calculates discount by multpying the percentage discount with the products price

 ## PremiumUserDiscount
 This Discount strategy prioritizes premium users by giving them a fixed discount of 20%


```python
'''
imports Abstract Base Class and abstractmethod
which controls the interface of other subclasses
ensuring that if subclasses lack certain methods defined under abstract methods are not present to raise an error
'''
from abc import ABC, abstractmethod

class Product:#Product instance
    def __init__(self, name: str, price: float) -> None:
        self.name = name
        self.price = price

    def __str__(self) -> str:
        return f'{self.name} - ${self.price}'
'''
The Abstract Base Class
Any class inheriting from it
Must have all methods defined within it
Otherwise an error will be raised
Methods under ABC'S
should not have an implementation
Each subclass should override or rather define its own specific methods
'''
class DiscountStrategy(ABC):
    @abstractmethod #officially declares a method as abstract(meaning it should be present in subclasses)
    def is_applicable(self, product: Product, user_tier: str) -> bool:
        pass

    @abstractmethod
    def apply_discount(self, product: Product) -> float:
        pass

class PercentageDiscount(DiscountStrategy):#subclass 1 inheriting from the ABC
    def __init__(self, percent: int) -> None:
        self.percent = percent

    def is_applicable(self, product: Product, user_tier: str) -> bool:#overrides the method in the ABC
        return self.percent <= 70# returns true when percentage is less or equal to 70
                                 #this prevents very high discounts

    def apply_discount(self, product: Product) -> float:#overrides the method in the ABC
        return product.price * (1 - self.percent / 100)#calculates percentage discount

class FixedAmountDiscount(DiscountStrategy):#subclass 2 inheriting from the ABC
    def __init__(self, amount: int) -> None:
        self.amount = amount

    def is_applicable(self, product: Product, user_tier: str) -> bool:
        return product.price * 0.9 > self.amount#returns true when 90% of the product is greater than the fixed amount

    def apply_discount(self, product: Product) -> float:
        return product.price - self.amount #calculates fixed amount discount

class PremiumUserDiscount(DiscountStrategy):#subclass 3 inheriting from the ABC
    def is_applicable(self, product: Product, user_tier: str) -> bool:
        return user_tier.lower() == 'premium'#returns true when user is premium user

    def apply_discount(self, product: Product) -> float:
        return product.price * 0.8#calculates 20% percentage discount for every premium users

'''
The Discount Engine is the main engine
That calculates the best price
from the various discount strategies
'''

class DiscountEngine:
    def __init__(self, strategies: list[DiscountStrategy]) -> None:
        self.strategies = strategies #a list of all the Discount Strategies

    def calculate_best_price(self, product: Product, user_tier: str) -> float:
        prices = [product.price]#esures when no discount strategy applies to a user, they pay the full amount

        for strategy in self.strategies:#loops through all strategies
            if strategy.is_applicable(product, user_tier):#validates them
                discounted = strategy.apply_discount(product)#calculates discount
                prices.append(discounted)

        return min(prices)#finds minimum price which translates to the best price
'''
Ensures when code is run in the main script it runs the main code
Otherwise it doesn't
'''
if __name__ == '__main__':
    product = Product('Wireless Mouse', 50.0)
    user_tier = 'Premium'

    strategies = [
        PercentageDiscount(10),
        FixedAmountDiscount(5),
        PremiumUserDiscount()
    ]

    engine = DiscountEngine(strategies)
    best_price = engine.calculate_best_price(product, user_tier)

print(f"Best price for {product.name} for {user_tier} user: ${best_price:.2f}")
```
