# Notebook Overview

This notebook demonstrates a selection sort algorithm function which sorts items in a list by repeatedly finding the smallest element from the unsorted portion of the list and swapping it with the first unsorted element. It begins by selecting the minimum value in the entire list and swapping it with the first element. Then it moves to the second position, finds the smallest value in the remaining unsorted elements, and swaps it with the second element. This process continues, moving through the list one element at a time, until the entire list is sorted.

# Breakdown
For this implementation i used python functions like min and copy to make it easier

1. create a copy of the origin list this is so as to create slices with the copy so a to allow modification of the main list
2. create a tracker index which iterates through each value
3. the first thing is to find the minimum value of the entire list then placing it at the first position
4. After minimum value is found from list its isolated/removed from list
5. increment tracker by 1 to modify the original list
6. Pass the edited list back to the function after removing the minimum value, repeat this process

````python
def selection_sort(array):
    i = 0#set tracker for main array
    edited_list = array.copy()
    while i< len(array):
        array[i] = min(edited_list)#finds minumum value in whole array
        edited_list.remove(array[i])
        i += 1





    return array
````

# Cons

Although the code works it has some major cons

1. Time complexity increases O(n^2) as ```min()```iterates through all values in a list

2. Removes(). Does shift items in memory from their original position
3. Copy() creates an O(n) time complexity and space complexity

4. remove() searching for the item in the list is an O(n) time complexity
