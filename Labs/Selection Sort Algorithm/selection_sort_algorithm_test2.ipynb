This is the second implementation of the selection sort Algorithm

```python
def selection_sort(array):
  i = 0
  while i < len(array):
    min_index = array.index(min(array[i:]),i) # starts search at specified position, index returned will be relative to the whole list
    if min_index == i:
      pass
    else:
      array[i],array[min_index] = array[min_index],array[i] #swaps the minimum value with the original position


    i += 1

  return array
```
