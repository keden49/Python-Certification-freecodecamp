This is a third attempt of the selection sort algorithm in this attempt i try to achieve a more efficient algorithm in terms of space and time complexities which my previous tests failed to achieve

# Goal Complexities

1. O(n2)- time
2. 0(1)-space

```python
def selection_sort(array):

  i = 0

  while i < len(array):
    min_index = i


    for j in range(i+1,len(array)):
      if array[j] < array[min_index]:
        array[j],array[min_index] = array[min_index],array[j] #swapping elements is unneccesary
    array[i] = array[min_index]
    i += 1

  return array







```
