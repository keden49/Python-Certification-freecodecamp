This is the 4th attempt to the selction sort algorithm, in my previous attempt I manage to achieve a time complexity of O(n^2) and space complexity of O(1) though some methods used were unneccesary as they used more effort than what was needed. In this attempt, I attempt to keep track of the minimum index value in the unsorted parts and if it changes to reassign it. Then at the end of the inner iteration to swap the true minimum value to the position of i(first position of the unsorted part of a list) and vice versa.

```python
def selection_sort(array):

  i = 0#keeps track of the main array passed

  while i < len(array):
    min_index = i


    for j in range(i+1,len(array)):#iterates through the unsorted list seraching for the minimum value
      if array[j] < array[min_index]:
        min_index = j
    array[i],array[min_index] = array[min_index],array[i]
    i += 1

  return array







```
