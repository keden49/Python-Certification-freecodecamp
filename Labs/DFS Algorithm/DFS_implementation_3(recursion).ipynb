"""
    Function used to illustrate the Depth First Search Algorithm Using Python

                  1
                 / \
                2   3
               / \   \
              4   5-> 6
                       \
                        7
    """

```python
graph = {1: [2,3],
         2: [4, 5],
         3: [6],
         4: [],
         5: [6],
         6: [7]
}


visited = []
queue = []
def dfs(visited, graph, val):

    if val not in visited:
        print(val)
        visited.append(val)#marks node as visited
        try:
            for neighbour in graph[val]:#loops through each node entry
                 dfs(visited, graph, neighbour)
        except:
            pass
dfs(visited, graph, 1)

```

# Conclusion

- Access a node
- Mark as visited
- Check its neighbours
- Repeat
