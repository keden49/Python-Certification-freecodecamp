```python
#couting number of elements
dictionary = {
    0: [1, 2],
    1: [2],
    2: [0, 3],
    3: [2]
}
n = len(dictionary)
```

```python
#creating 2D matrix all initialized with zeros

#matrix = [[0]*n]*n # creates reference to the same inner list
#print(matrix)

matrix = [[0 for _ in range(n)] for _ in range(n)]
print(matrix)
```

Tests:

```python
def adjacency_list_to_matrix(dictionary):#Test 1,2

    #counting number of items in dictionary
    n = len(dictionary)#Test 3

    #intializing matrix with zeros

    matrix = [[0 for _ in range(n)] for _ in range(n)]

    #replacing edges with 1
    for key,value in dictionary.items():

        #accessing specific edges

        for i in range(len(value)):
            matrix[key][value[i]] = 1 #Test 4
        print(matrix[key])# Test 5

    return matrix#Test 6




```
