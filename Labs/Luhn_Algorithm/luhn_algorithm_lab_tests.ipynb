1. You should have a function named verify_card_number.
2. verify_card_number('453914889') should return VALID!.
3. verify_card_number('4111-1111-1111-1111') should return VALID!.
4. verify_card_number('453914881') should return INVALID!.
5. verify_card_number('1234 5678 9012 3456') should return INVALID!.
6. verify_card_number should return VALID! when called with a valid credit card number.
7. verify_card_number should return INVALID! when called with an invalid credit card number.

```python
def verify_card_number(card_number):#test 1

    #Removing '-' and ' '
    card_number = card_number.replace('-','').replace(' ','')

    '''reversing string to enable right to left movement'''

    reverse_card_number = card_number[::-1]

    '''excluding check point(beggining of odd numbers which will not be doubled)
    since i reversed the string the checkpoint is at the first position'''

    odd_digits = reverse_card_number[0::2]#reprsent digits that will not be doubled
    even_digits = reverse_card_number[1::2]#reprsent digits that will be doubled
    sum_of_odd_digits = 0
    sum_of_even_digits = 0

    for i in odd_digits:
        sum_of_odd_digits += int(i)#convert string to integer

    for n in even_digits:
        n = int(n) * 2


        '''summing digits of numbers whih when doubled result in a number greater than 10 i.e numbers above 5'''

        if n >= 10:
            n = (n//10) + (n%10)

        sum_of_even_digits += n

    total = sum_of_odd_digits + sum_of_even_digits

    '''checking if total is a multiple of 10 meaning its remainder should always be 0'''

    if total % 10 == 0:
        return 'VALID!'

    else:
        return 'INVALID!'


print(verify_card_number('1234 5678 9012 3456'))


```
