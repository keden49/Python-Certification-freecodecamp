```python
class GameCharacter:
    def __init__ (self,name):
        self._name = name
        self.health = 100
        self.mana = 50
        self._level = 1

    # defining name getter
    @property
    def name(self):
        return self._name

    #health properties
    @property
    def health(self):
        return self._health

    @health.setter
    def health(self,value):
        #capping value from 0 to 100
        if value < 0:
            self._health = 0
        elif value > 100:
            self._health = 100
        else:
            self._health = value

    #setting mana properties

    @property
    def mana(self):
        return self._mana

    @mana.setter
    def mana(self,value):
        if value < 0:
            self._mana = 0
        elif value > 50:
            self._mana = 50
        else:
            self._mana = value

    #setting level getter

    @property
    def level(self):
        return self._level


    def level_up(self):
        self._level += 1
        self.health = 100
        self.mana  = 50

        print(f"{self.name} leveled up to {self.level}!")

    def __str__ (self):
        return(f"Name: {self.name}\nLevel: {self.level}\nHealth: {self.health}\nMana: {self.mana}")

hero = GameCharacter('Kratos') # Creates a new character named Kratos
print(hero)










```

```python
hero.level_up()
```
