#Notebook Overview

This Notebook aims to build the quick sort algolrithm on top of the merge sort algorithm

```python
def quick_sort(array):
  if len(array) <= 1:#starts at single elements
    return array

  mid = len(array)//2
  left_array = quick_sort(array[:mid])
  right_array= quick_sort(array[mid:])

  pivot_point = array[-1]#setting comparison integer as the last item in a list

  l = 0 #tracks left_array array
  r = 0 #tracks right_array array

  less_int = []#stores integers less than the pivot point
  equal_int = []#stores integers equal than the pivot point
  great_int = []#stores integers greater than the pivot point

  #ensures loop runs only when elements are present

  while l< len(left_array) and r < len(right_array):
    #sorting
    if left_array[l] < right_array[r]:
      if pivot_point > left_array[l]:
        less_int.append(left_array[l])
      elif pivot_point < left_array[l]:
        great_int.append(left_array[l])
      else:
        equal_int.append(left_array[l])
      l += 1


    else:
      if pivot_point > right_array[r]:
        less_int.append(right_array[r])
      elif pivot_point < right_array[r]:
        great_int.append(right_array[r])
      else:
        equal_int.append(right_array[r])
      r += 1


  sorted_list = less_int + equal_int + great_int
  sorted_list.extend(left_array[l:])
  sorted_list.extend(right_array[r:])


  return sorted_list





```
