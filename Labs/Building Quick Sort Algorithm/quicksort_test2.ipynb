```python
# placing pivot at correct position in original array
# Where elements to its left are lesser than pivot and elements to its right are greater


def quick_sort(array):

  #stop recursion

  if len(array) <= 1:
    return


  pivot = array[-1]#set last element as element

  i = -1 #traces lesser elements
  j = 0 #moves through array
  #temp = [] #temporary position enables swapping
#only runs when they are items still present in the list
  while j < len(array):

    if array[j] <= pivot:
      i += 1
      temp = array[i]#stores temporary values
      #swapping values
      array[i] = array[j]
      array[j] = temp
      j += 1

    else:
       j += 1 #ignores the value if it is larger than pivot

```

# Conclusion

Final array has all elements greater than it to the right and all elements lesser than it to the left

```python
print(quick_sort([8,2,4,7,1,3,9,5]))
```
