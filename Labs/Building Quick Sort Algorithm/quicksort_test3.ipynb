```python
def quick_sort(arr):
  if len(arr) <= 1:
    return arr #stops recursion

  pivot = arr.pop()
  left_arr = []
  right_arr = []

  for x in arr:
    if x <= pivot:
      left_arr.append(x)
    else:
      right_arr.append(x)

  left_arr = quick_sort(left_arr)#always will return a sorted list
  right_arr = quick_sort(right_arr)#always will return a sorted list

  return left_arr + [pivot]+ right_arr

```

```python
quick_sort([2, 4, 1, 3, 5, 7, 6, 9, 8])
```
