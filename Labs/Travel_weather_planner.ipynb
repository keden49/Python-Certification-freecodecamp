# Objective

For this lab, you will use conditional statements to determine whether commuting is possible based on the weather, the distance to travel, and the availability of a vehicle

# Key Concepts Learned
1. Conditional statements
2. Boolean Operators
3. Boolean Manipulation

```python
distance_mi = 7
is_raining = False
has_bike = True
has_car = True
has_ride_share_app = True


if not distance_mi:
    print(True)


elif distance_mi <= 1 and not is_raining:
    print(True)


elif 1 < distance_mi <= 6 and all([has_bike, not is_raining]):
    print(True)


elif distance_mi > 6 and any([has_ride_share_app, has_car]):
  print(True)


else:
    print(False)
```
