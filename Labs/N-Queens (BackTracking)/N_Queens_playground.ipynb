# Logic

Backtracking is an algorithmic technique for solving problems incrementally by trying possibilities and abandoning them ("backtracking") as soon as they are determined to be invalid or lead to a dead end. It builds candidates for solutions, typically via recursion, and reverts to the previous decision point when a path fails, often used to find all solutions or the best solution to constraint satisfaction problems

# Things to keep in mind

Queens should not be in same column,row, any of the diagonals

# Positive Diagonals

For Queens in same "+" diagonals, row + column values are the same

# Negative diagonals

For Queens in "-" diagonals , row - column values are the same

```python
def dfs_n_queens(n):
    #initializing chessboard
    chess_board = [["." for _ in range(n)] for _ in range(n)]

    '''
    tracking of columns,positive diagonals,negative_diagonals to avoid placement of queens in similar positions
    '''
    column = []
    pos_diag = []
    neg_diag = []
    results = [] #list to store solutions

    '''
    Fn that makes changes to the chess board
    forward and backward in time
    '''
    def backtrack(row):

        #keep track of main variables
        nonlocal column,pos_diag,neg_diag,chess_board

        # base case

        if row == n:
            solution = [r.index('Q') for r in chess_board]
            results.append(solution)
            return

        # finding queens placement

        for col in range(n):

            #avoids placing queen attacking zones
            if col in column or (row + col) in pos_diag or (row-col) in neg_diag:
                continue


            #updating positions

            column.append(col)
            pos_diag.append(row + col)
            neg_diag.append(row - col)
            chess_board[row][col] = 'Q'

            '''
            recursive call
            fn looks fwd in time and makes decision based on that
            '''
            backtrack(row + 1)

            '''
            based on what its seen in the future
            if it will meets deadends(where future queen has no positions)
            it changes its current position
            to achieve that, current changes are removed
            '''

            column.remove(col)
            pos_diag.remove(row + col)
            neg_diag.remove(row - col)
            chess_board[row][col] = '.'
    backtrack(0)


    return results




```

# Flow


1. Initialize an empty chess board
2. Track all attack zones, columns, and diagonals(both positive and negative)
3. Define a backtracking function for testing out possibilities and abandoning them if they lead to invalid dead ends
4. Place queens on appropiate positions(attack freezones)
5. Change positions once it meets a dead end
6. Make decisions by looking forward in time
