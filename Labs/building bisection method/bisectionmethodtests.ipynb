```python
#Test 1
def square_root_bisection(target,tolerance = 0.001,max_iter = 100):#Test 2,3,4
    if target < 0: #capping negative values
        raise ValueError("Square root of negative number is not defined in real numbers")#Test 5
    elif target in [1,0]:
        print(f"The square root of {target} is {target}")
        return target
    else:
        '''
        sets upper bound to target if numberis above 1
        else for numbers between 1 and 0
        whose square roots maybe larger than the target itself eg(0.25) root = 0.5
        '''
        upper_bound = max(1,target)
        lower_bound = 0
        root = None

        for _ in range(max_iter):
            '''
            begginning of each iteration a new midpoint is set
            depending on the lower and upper bounds
            '''
            mid = (upper_bound+lower_bound)/2
            square_mid = mid**2 #helps identify tolerance value
            
            '''
             set range within the tolerance
             setting the tolerance within the boundaries
             ensures that the the predicted midpoint
             within that tolerance range is such that
             the true root will also be within that very range
             from the approximated root
             '''


            if (upper_bound-lower_bound) < tolerance:
                root = mid
                break


                '''


            when the square of thr midpoint is above the square target
            the function loop shifts to the lower half
            by setting the upper bound to the midpoint the new midpoint will be lower therefore the square mid reduces

                '''


            elif square_mid > target:
                upper_bound = mid

            else:
                #raises lower_bound when squaremid is below target
                lower_bound = mid

#square_root_bisection(0)
        if root is None:
          print(f"Failed to converge within {max_iter} iterations")
          return None

        else:
          print(f"The square root of {target} is approximately {root}")
          return root
```
