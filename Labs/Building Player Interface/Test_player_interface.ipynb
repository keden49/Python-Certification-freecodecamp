```python
from abc import ABC,abstractmethod
import random

class Player(ABC):#Test 1,2
    def __init__(self):#Test 3,4
        self.moves = []#Test 5
        self.position = (0, 0)#Test 6
        self.path = [self.position]#Test 7

    def make_move(self):#Test 8,9
        #generating random choice

        random_move = random.choice(self.moves)# Test 10
        #setting new position
        new_position = tuple([random_move[i]+self.position[i] for i in range(2)])#Test 11
        self.position = new_position#Test 12
        #Appending new position to path
        self.path.append(new_position)#Test 12
        return self.position #returns new position #Test 13


    @abstractmethod#Test 16
    def level_up(self):#Test 14,15
        pass #Test 17


#creating the Pawn class

class Pawn(Player):#Test 18,19
    def __init__(self):#Test 20,21
        super().__init__()#calls on parents constructor # Test 22
        self.moves = [(0,1),(0,-1),(1,0),(-1,0)]# Test 23


    def level_up(self):#Test 24
        diagonal_moves = [(-1,-1),(-1,1),(1,1),(1,-1)]
        self.moves.extend(diagonal_moves)# Test 25


Pawn()# An instance of pawn







```
