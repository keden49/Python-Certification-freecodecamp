## Code Logic
I started by defining the abstract class Player which contains initial attributes like
```moves``` - which represents all the positions a pawn in a chess game can make
```position``` - refelcts the current cordinate postion of a pawn on a chess board
```path``` - tracks pawns movements through the chess board

The ```make move``` method as the name suggests allows one to make a move with he pawn, it calculates a random location a pawn can move. Then computes the new location by adding the prior and new coordinate.

Then I defined the main Pawn Object, which is what moves. The object overrides the parents(Player) init method. This is so as to set the cardinal positions a pawn can move in the moves attribute this is up(0,1),down(0,-1),left(-1,0) and right positions(1,0)

The method still needs references to the parents class init method so as to set the original pawn position to (0,0) and path to start tracking.

Pawn also overrides the level up method and this method allows, diagonal movements which in a typical chess game is how a pawn can defeat a neighbouring pawn or seek defence and with that the interface is complete.

# Key Highlights of this Code:

Encapsulation: All data (position, moves) is kept inside the object.

Inheritance: Pawn gets all the logic from Player for free.

Polymorphism: The level_up method allows different characters to grow in unique

````python
from abc import ABC, abstractmethod
import random

class Player(ABC):
    def __init__(self):
        self.moves = []
        self.position = (0, 0)
        self.path = [self.position]

    def make_move(self):
        random_move = random.choice(self.moves)

        # Setting new position
        new_position = tuple([random_move[i] + self.position[i] for i in range(2)])
        self.position = new_position

        # Appending new position to path
        self.path.append(new_position)
        return self.position

    @abstractmethod
    def level_up(self):
        pass

class Pawn(Player):
    def __init__(self):
        super().__init__()
        self.moves = [(0, 1), (0, -1), (1, 0), (-1, 0)]

    def level_up(self):
        diagonal_moves = [(-1, -1), (-1, 1), (1, 1), (1, -1)]
        self.moves.extend(diagonal_moves)

pawn = Pawn()
````

# New Implementation

Unpacking tuple elements and summing them up

````python
from abc import ABC, abstractmethod
import random

class Player(ABC):
    def __init__(self):
        self.moves = []
        self.position = (0, 0)
        self.path = [self.position]

    def make_move(self):
        dx,dy = random.choice(self.moves)

        # Setting new position
        x,y = self.position
        new_position = (x + dx, y + dy)
        self.position = new_position

        # Appending new position to path
        self.path.append(new_position)
        return self.position

    @abstractmethod
    def level_up(self):
        pass

class Pawn(Player):
    def __init__(self):
        super().__init__()
        self.moves = [(0, 1), (0, -1), (1, 0), (-1, 0)]

    def level_up(self):
        diagonal_moves = [(-1, -1), (-1, 1), (1, 1), (1, -1)]
        self.moves.extend(diagonal_moves)

pawn = Pawn()
````
