```python
class Planet: #creating empty class object
    def __init__(self,name,planet_type,star):#initializing instance
        if not all(isinstance(entry,str) for entry in [name,planet_type,star]):
            raise TypeError(f'name, planet type, and star must be strings')

        #cehck if all variables are non-empty
        elif not all([name,planet_type,star]):
            raise ValueError('name, planet_type, and star must be non-empty strings')
        else:
            self.name = name
            self.planet_type = planet_type
            self.star = star

    def orbit(self):#defining orbit method
        return f'{self.name} is orbiting around {self.star}...'

    def __str__(self):
        return f'Planet: {self.name} | Type: {self.planet_type} | Star: {self.star}'


planet_1 = Planet('Earth','Terrestrial','Sun')
planet_2 = Planet('Jupiter','Gas','Sun')
planet_3 = Planet('Neptune','Ice Giant','Sun')
#Print __str__ method
print(planet_1)
print(planet_2)
print(planet_3)

#PrintOrbit Method
print(planet_1.orbit())
print(planet_2.orbit())
print(planet_3.orbit())



#Object Location In Memory
print(hex(id(planet_1)))
print(hex(id(planet_2)))
print(hex(id(planet_3)))
```
