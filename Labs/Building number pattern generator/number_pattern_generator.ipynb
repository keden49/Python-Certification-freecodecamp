```python
def number_pattern(n):
    non_spaced_pattern = []
    if not isinstance(n,int):
        return 'Argument must be an integer value.'
    if n < 1:
        return 'Argument must be an integer greater than 0.'



    for i in range(n):
        enumerated_number= i + 1
        non_spaced_pattern.append(str(enumerated_number))
        spaced_pattern= ' '.join(non_spaced_pattern)

    return spaced_pattern


input_1 = 4
print(number_pattern(input_1))

input_2= 12
print(number_pattern(input_2))
```
