```python
full_dot = '●'
empty_dot = '○'
#creating function
def create_character(name, strength, intelligence, charisma):
  #Validate name is a string
    if not isinstance(name, str):
        return 'The character name should be a string'
  #Validate name is not empty
    if name == '':
        return 'The character should have a name'
  #Validate name is =< 10
    if len(name) > 10:
        return 'The character name is too long'
  #Validate no space in name
    if ' ' in name:
        return 'The character name should not contain spaces'
  #list of characters statistics
    stats = [strength, intelligence, charisma]
  #loop through stats
    for stat in stats:
        if not isinstance(stat, int):#Validates all stats are intergers
            return 'All stats should be integers'
        if stat < 1:#Validates stats are => 1
            return 'All stats should be no less than 1'
        if stat > 4:#Validates stats are =< 4
            return 'All stats should be no more than 4'

    if sum(stats) != 7:#Validates sum of stats is = 7
        return 'The character should start with 7 points'
   # Create visual stat bars
    S = (full_dot) * strength + empty_dot * (10 - strength)
    I = (full_dot) * intelligence + empty_dot * (10 - intelligence)
    C = (full_dot) * charisma + empty_dot * (10 - charisma)
   # returns output
    return f'{name}\nSTR {S}\nINT {I}\nCHA {C}'

RPG = create_character('FATHELA', 4, 2, 1)
print(RPG)
```
